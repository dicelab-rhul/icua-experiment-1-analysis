```python
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
import numpy as np

# Example data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a set of line segments so that we can color them individually
# This creates the points as a N x 1 x 2 array so that we can stack points together easily to get the segments.
points = np.array([x, y]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)

# Create a boolean array where True will be one color, and False will be another
colors = np.sin(x) > 0

# Map the boolean array to a color array
cmap = np.array(['red', 'blue'])
line_colors = cmap[colors.astype(int)]

# Create a LineCollection from the segments
lc = LineCollection(segments, colors=line_colors, linewidth=2)

# Create the plot
fig, ax = plt.subplots()
ax.add_collection(lc)
ax.autoscale()
ax.set_xlim(x.min(), x.max())
ax.set_ylim(y.min(), y.max())
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm

# Sample data for two histograms
data1 = np.random.randn(100) * 10
data2 = np.random.randn(100)

# Find the global minimum and maximum values to normalize the color map
global_min = min(data1.min(), data2.min())
global_max = max(data1.max(), data2.max())

# Create a colormap
cmap = cm.viridis

# Create the subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Function to plot a histogram where color of each bar represents its x-value
def plot_color_histogram(ax, data, cmap, vmin, vmax):
    # Compute the histogram data
    counts, bins = np.histogram(data, bins=10)
    bin_centers = 0.5 * (bins[:-1] + bins[1:])
    
    # Normalize the bin centers to the global range of the data
    norm = plt.Normalize(vmin, vmax)
    
    # Map the bin centers to the colormap
    colors = cmap(norm(bin_centers))
    
    # Plot each bar individually
    for c, p, color in zip(counts, bins, colors):
        ax.bar(p, c, width=bins[1] - bins[0], color=color, edgecolor='grey')

    x_range = bins[-1] - bins[0]
    y_range = counts.max()
    ax.set_aspect(aspect=x_range/y_range)
      
# Plot histograms with colored bars using the global color scale
plot_color_histogram(axes[0], data1, cmap, global_min, global_max)
plot_color_histogram(axes[1], data2, cmap, global_min, global_max)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.colors import LinearSegmentedColormap
# Load the viridis colormap
viridis = matplotlib.colormaps.get_cmap('viridis')

# Create a custom colormap that changes strongly at a threshold
def create_custom_cmap(cmap, threshold):
    colors = cmap(np.linspace(0, 1, cmap.N))
    
    print(colors.shape)
    initial_color = np.tile(colors[0], (int(cmap.N/threshold),1))
    print(initial_color.shape)
    
    # Create a new list of colors with a strong change at the threshold
    new_colors = np.vstack((initial_color,
                           colors, 
                        ))
    print(new_colors)
    # Create a new LinearSegmentedColormap
    new_cmap = LinearSegmentedColormap.from_list('CustomCmap', new_colors)
    return new_cmap

# Define the threshold
threshold = 10  # This could be any value between 0 and 1

# Create the custom colormap
custom_cmap = create_custom_cmap(viridis, threshold)

# Use the custom colormap in a plot
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x, y, c=x, cmap=custom_cmap)
plt.colorbar()
plt.show()
```
