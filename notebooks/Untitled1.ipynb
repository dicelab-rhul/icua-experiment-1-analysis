```python

import icua_analysis as ia
import numpy as np
import matplotlib.pyplot as plt

THRESHOLD = 200
dt = 0.016

for (par, trial), data in ia.load_tabularised().has_eyetracking():
    df = data['eyetracking_data'].loc[:100]
    
    
    d = np.sqrt((np.diff(df['x'].to_numpy()) / dt) ** 2 + (np.diff(df['y'].to_numpy()) / dt)  ** 2)
    
    #__d = np.sqrt(np.diff(df['x'].to_numpy()) ** 2 + np.diff(df['y'].to_numpy()) ** 2) / (0.016)
    
    print(df.columns)
    plt.figure(figsize=(20,1))
    plt.plot(1 + np.arange(len(d)), d)
    plt.plot(np.arange(len(df['gaze'])), df['gaze'].to_numpy() * d.max())
    
    plt.hlines(THRESHOLD, 0, len(d), color='red')
    
    break
```
